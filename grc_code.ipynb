Notes:
1. Some outliers in the data are possible to be actual values than being bad data. For the consideration of this exercise, any extreme value was considered a bad data. For the most part, after some research on the terms, most of the extreme values did turn out to be bad data. 
2. For most columns that were manipulated were created as new fields, with "Calc" as a prefix, to retain actual values.
3. Answers to the questions are calculated based on the new/manipulated fields wherever applicable. If a different imputation/manipulation method was used for the data in the original solution, the answer might be different. 
4. Before performing data exploration in python, the csv file was also throughly examined in excel.

```python
#Importing libraries and packages

from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
import pandas as pd
from datetime import datetime
```

```python
#Importing dataset and data exploration

dataset = pd.read_csv("grc_qi_coding_exercise.csv")
print(dataset.head(10))
dataset.describe()
dataset.dtypes
```

Clearly there are some outliers in the data and it needs to be cleaned. 

```python
#Counting unique patients 

dataset['PatID'].nunique()
```

```python
pd.unique(dataset.SiteName)

#Replacing the typo with correct site 
dataset['SiteName'] = dataset.SiteName.replace('Clinc 2', 'Clinic 2')

#Validating
pd.unique(dataset.SiteName)
```

```python
#Calcuating age at the time of the visit since there is some bad data
dataset['VisitDate'] = pd.to_datetime(dataset.VisitDate, format='%m/%d/%Y')
dataset['DOB'] = pd.to_datetime(dataset.DOB, format='%m/%d/%Y')

#Age calculated by (VisitDate - DOB)
dataset['CalcAgeAtVisit'] = (dataset.VisitDate - dataset.DOB).astype('<m8[Y]').astype(int)
```

```python
#Comparing the calcuated and given age to decide the best method for data imputation
dataset.loc[~(dataset['AgeAtVisit'] == dataset['CalcAgeAtVisit'])]
```

```python
#Creating a final calculated field for Age to use original age that was given in the dataset if the calulated age is 0. Some of the patients have a visit date same as DOB, so it seems that visit date could be entered wrong as these patients have other visits too. 
dataset['CalcAge'] = dataset.AgeAtVisit
dataset.loc[dataset.AgeAtVisit == 0, 'CalcAge'] = dataset.CalcAgeAtVisit

#Validating imputation 
dataset.loc[~(dataset['AgeAtVisit'] == dataset['CalcAgeAtVisit'])]
```

```python
#Converting Sex to uppercase

pd.unique(dataset.Sex)
dataset['CalcSex'] = dataset.Sex.str.upper()
pd.unique(dataset.CalcSex)
```

```python
#Replacing outliers in A1C with mode

dataset['CalcA1C'] = dataset.A1C.replace(99.9, dataset.A1C.mode()[0])
dataset[dataset.A1C == 99.9]
```

```python
#Replacing outliers in SysBP & DiasBP with mode

dataset['CalcSysBP'] = dataset.SysBP.replace(999, dataset.SysBP.mode()[0])
dataset[dataset.SysBP == 999]

dataset['CalcDiasBP'] = dataset.DiasBP.replace(999, dataset.DiasBP.mode()[0])
dataset[dataset.DiasBP == 999]
```

```python
#Mode was imputed for the patients on whom the depression screening was administered but had nulls in PHQ score
dataset.loc[(dataset.DepScreen == 'Y') & (pd.isnull(dataset.PHQ)), 'PHQ'] = dataset.PHQ.mode()[0]

#Filling other nulls with -1 since they weren't administered for depression screening
dataset.PHQ.fillna(-1, inplace=True)

#Validation
dataset.PHQ.value_counts()
```

```python
#Imputing N/A for DepFollowUpSch for the patients that didn't have a PHQ score over 4 to only have Y or N for patients who were actually eligible for the follow-up
dataset['CalcDepFollowUpSch'] = dataset['DepFollowUpSch'] 
dataset.loc[(dataset.PHQ <= 4), 'CalcDepFollowUpSch'] = "N/A"

#Validation
dataset.CalcDepFollowUpSch.value_counts()

#Clean Dataset
dataset
```

Now the data is clean, we find the answers to the questions. 

Q1: How many patients visited Clinic 1 on or before April 30, 2030?

Ans: 32 unique patients (44 with duplicates)

```python
#Creating a date variable for comparison
DateComp = datetime(2030, 4, 30)
dataset[(dataset.SiteName == "Clinic 1") & (dataset.VisitDate <= DateComp)].PatID.count()
dataset[(dataset.SiteName == "Clinic 1") & (dataset.VisitDate <= DateComp)].PatID.nunique()
```

Q2: What was the average patient age among all visits with an A1C value of greater
than or equal to 9.0?

Ans: 54 was the average age. 

Note that this calculation uses the "CalcAge" & "CalcA1C" instead of the original fields, so the solution can vary. Since original fields had bad data, this would be a more accurate representation.

```python
#Calcating mean
dataset[dataset.CalcA1C >= 9].CalcAge.mean()
```

Q3: How many patients aged 50 or younger on the visit date had recorded blood pressures consistent with hypertension (systolic BP > 140 AND diastolic BP > 90) on at least two separate visits?

Ans: 35 patients

Note that since outliers in the SysBP and DiasBP were replaced by mode, the original values weren't used in this calculation, but the cleaned fields.  

```python
#Filtering for the given specs 
PatientsTemp = dataset[(dataset.CalcAge <= 50) & (dataset.CalcSysBP > 140) & (dataset.CalcDiasBP > 90)]

#Counting occurances
PatientsTemp['Count'] = PatientsTemp.groupby('PatID')['PatID'].transform('count')
PatientsTemp

#Counting unique patients
PatientsTemp[PatientsTemp.Count >=2].PatID.nunique()
```

Q4: How many patients at Clinic 2 received a depression screening and had a PHQscore of 4 or lower?

Ans: 13 unique patients (24 with duplicates)

```python
dataset[(dataset.SiteName == "Clinic 2") & (dataset.DepScreen == "Y") & (dataset.PHQ <= 4)].PatID.count()
dataset[(dataset.SiteName == "Clinic 2") & (dataset.DepScreen == "Y") & (dataset.PHQ <= 4)].PatID.nunique()
```

Q5: Among male patients with hypertension who visited Clinic 1 or Clinic 3, what was the proportion of visits at each clinic during which a follow-up visit for hypertension was scheduled?

Ans: Clinic 1 = 19/41 and Clinic 3 = 20/34

```python
#Filtering for the given specs 
MaleTemp = dataset[(dataset.CalcSex == "M") & (dataset.CalcSysBP > 140) & (dataset.CalcDiasBP > 90) & ((dataset.SiteName == "Clinic 1")|(dataset.SiteName == "Clinic 3"))]

#Proportion of visits for Clinic 1
(MaleTemp[(MaleTemp.SiteName == "Clinic 1") & (MaleTemp.HypFollowUpSch == "Y")].PatID.count())
(MaleTemp[MaleTemp.SiteName == "Clinic 1"].PatID.count())

#Proportion of visits for Clinic 3
(MaleTemp[(MaleTemp.SiteName == "Clinic 3") & (MaleTemp.HypFollowUpSch == "Y")].PatID.count())
(MaleTemp[MaleTemp.SiteName == "Clinic 3"].PatID.count())
```
